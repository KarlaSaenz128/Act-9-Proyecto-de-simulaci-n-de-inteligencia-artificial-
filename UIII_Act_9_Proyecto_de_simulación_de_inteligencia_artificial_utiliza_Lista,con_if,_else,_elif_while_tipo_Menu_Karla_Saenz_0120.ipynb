```python
influencers = []

while True:
    print("\n===== MENU INFLUENCER =====")
    print("1. Agregar influencer")
    print("2. Leer influencer")
    print("3. Listar influencers")
    print("4. Buscar influencer")
    print("5. Modificar influencer")
    print("6. Borrar influencer")
    print("7. Salir")

    opcion = input("Selecciona una opción: ")

    # 1. Agregar influencer
    if opcion == "1":
        nombre = input("Escribe el nombre del influencer: ")
        influencers.append(nombre)
        print("Influencer agregado correctamente")

    # 2. Leer influencer
    elif opcion == "2":
        if len(influencers) == 0:
            print("La lista está vacía")
        else:
            try:
                posicion = int(input("Número de posición: "))
                if posicion >= 0 and posicion < len(influencers):
                    print("Influencer encontrado:", influencers[posicion])
                else:
                    print("Posición incorrecta")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la posición.")

    # 3. Listar influencers
    elif opcion == "3":
        if len(influencers) == 0:
            print("No hay influencers registrados")
        else:
            i = 0
            while i < len(influencers):
                print(i, "-", influencers[i])
                i = i + 1

    # 4. Buscar influencer
    elif opcion == "4":
        buscar = input("Nombre del influencer a buscar: ")
        if buscar in influencers:
            print("Sí existe el influencer")
        else:
            print("No se encontró el influencer")

    # 5. Modificar influencer
    elif opcion == "5":
        if len(influencers) == 0:
            print("La lista está vacía")
        else:
            try:
                posicion = int(input("Posición a modificar: "))
                if posicion >= 0 and posicion < len(influencers):
                    nuevo = input("Nuevo nombre: ")
                    influencers[posicion] = nuevo
                    print("Influencer modificado")
                else:
                    print("Posición incorrecta")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la posición.")

    # 6. Borrar influencer
    elif opcion == "6":
        if len(influencers) == 0:
            print("La lista está vacía")
        else:
            try:
                posicion = int(input("Posición a borrar: "))
                if posicion >= 0 and posicion < len(influencers):
                    eliminado = influencers.pop(posicion)
                    print("Influencer eliminado:", eliminado)
                else:
                    print("Posición incorrecta")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la posición.")

    # 7. Salir
    elif opcion == "7":
        print("/////Lista/////")
        print("\nDatos guardados correctamente")
        print("Programa terminado")
        print("Programa realizado por Karla Saenz NL = 52")
        break

    # Opción inválida
    else:
        print("Opción no válida")
```
