```python
# --- SIMULADOR DE IA PARA INFLUENCERS ---

# Creamos la lista de influencers inicialmente vacía
lista_influencers = []

# Variable para controlar el ciclo principal
continuar = True

print("==================================================")
print(" 🤖 BIENVENIDO AL IA-INFLUENCER MANAGER v1.0 🤖 ")
print("      Simulador de Inteligencia Artificial        ")
print("==================================================")

while continuar:
    # Mostramos el menú interactivo
    print("\n--- MENÚ DE OPCIONES ---")
    print("1. Registrar nuevo influencer")
    print("2. Ver lista de influencers estrella")
    print("3. Ejecutar Diagnóstico de IA (Predicción de éxito)")
    print("4. Salir del programa")

    opcion = input("\nSelecciona una opción (1-4): ")

    # OPCIÓN 1: Registrar Influencer
    if opcion == "1":
        nombre = input("Ingresa el nombre del influencer: ")
        # Agregamos el nombre a nuestra lista
        lista_influencers.append(nombre)
        print("¡" + nombre + " ha sido registrado con éxito!")

    # OPCIÓN 2: Ver la lista
    elif opcion == "2":
        print("\n--- LISTA DE INFLUENCERS REGISTRADOS ---")
        # Validamos si la lista está vacía con un simple if
        if lista_influencers == []:
            print("La lista está vacía. ¡Registra a alguien primero!")
        else:
            print(lista_influencers)

    # OPCIÓN 3: El "Cerebro" de la IA (Lógica de predicción)
    elif opcion == "3":
        print("\n--- DIAGNÓSTICO INTELIGENTE (IA) ---")

        if lista_influencers == []:
            print("🤖 IA: No hay influencers en la base de datos para analizar.")
        else:
            nombre_analizar = input("¿A qué influencer de la lista deseas analizar?: ")

            # Verificamos si el influencer existe en la lista
            if nombre_analizar in lista_influencers:
                print("\nResponde las siguientes preguntas para que la IA prediga su futuro:")
                seguidores = int(input("¿Cuántos seguidores tiene? (ej. 50000): "))
                likes_promedio = int(input("¿Cuántos likes promedio tiene por post?: "))
                hater_level = input("¿Tiene muchos haters? (si/no): ")

                # La IA calcula la tasa de interacción (engagement) de forma simple
                interaccion = likes_promedio / seguidores

                print("\n🤖 PROCESANDO DATOS POR LA IA...")

                # Sistema de toma de decisiones (Algoritmo de la IA)
                if interaccion > 0.20 and hater_level == "no":
                    print("🔮 PREDICCIÓN: ¡" + nombre_analizar + " será una superestrella mundial! Su comunidad es muy fiel.")
                elif interaccion > 0.10 and hater_level == "si":
                    print("🔮 PREDICCIÓN: " + nombre_analizar + " tendrá mucha polémica, pero ganará dinero por vistas. ¡Cuidado con la cancelación!")
                elif interaccion <= 0.10 and hater_level == "no":
                    print("🔮 PREDICCIÓN: Contenido seguro pero aburrido. La IA recomienda cambiar de nicho o subir más tiktoks.")
                elif interaccion <= 0.10 and hater_level == "si":
                    print("🔮 PREDICCIÓN: Alerta roja. Pocos likes y muchos haters. ¡Peligro de declive total!")
                else:
                    print("🔮 PREDICCIÓN: Datos inestables. ¡Sigue creando contenido!")
            else:
                print("❌ Ese influencer no está registrado en la lista.")

    # OPCIÓN 4: Salir
    elif opcion == "4":
        print("\n🤖 IA: Apagando sistema... ¡Suerte en el algoritmo!")
        # Cambiamos la condición a False para romper el while
        continuar = False

    # Validar si el usuario mete una opción que no existe
    else:
        print("❌ Opción inválida. Por favor, elige un número del 1 al 4.")

print("\n==================================================")
print("       Programa terminado. ¡Gracias! 🔥           ")
print("==================================================")
```
