```python
# Simulación de IA para Influencers
# La lista de seguidores inicia vacía
seguidores = []
opcion = ""

print("--- BIENVENIDO AL SISTEMA DE IA PARA INFLUENCERS ---")

# Ciclo principal para mantener el menú activo
while opcion != "5":
    print("\n¿Qué deseas hacer hoy?")
    print("1. Agregar un nuevo seguidor (Grow)")
    print("2. Ver mi lista de comunidad")
    print("3. Consultar a la IA: ¿Cómo va mi cuenta?")
    print("4. Eliminar último seguidor (Hater)")
    print("5. Salir")

    opcion = input("\nElige una opción (1-5): ")

    if opcion == "1":
        # Se solicita el nombre y se agrega a la lista usando append
        nuevo = input("Ingresa el nombre del nuevo seguidor: ")
        seguidores.append(nuevo)
        print("¡IA: Genial! Tu comunidad está creciendo.")

    elif opcion == "2":
        # Se verifica si la lista tiene elementos con if
        if len(seguidores) == 0:
            print("IA: Tu lista está vacía. ¡Empieza a crear contenido!")
        else:
            print("Tu comunidad actual es:")
            # Recorrido manual con while en lugar de ciclo for
            indice = 0
            while indice < len(seguidores):
                print("- " + seguidores[indice])
                indice = indice + 1

    elif opcion == "3":
        # La IA analiza el tamaño de la lista con elif
        cantidad = len(seguidores)
        if cantidad == 0:
            print("IA: Análisis... Necesitas publicar tu primer video hoy.")
        elif cantidad < 3:
            print("IA: Análisis... Vas por buen camino, sigue así.")
        else:
            print("IA: Análisis... ¡Eres tendencia! Tu comunidad es sólida.")

    elif opcion == "4":
        # Se elimina el último elemento si la lista no está vacía
        if len(seguidores) > 0:
            eliminado = seguidores.pop()
            print("IA: Se ha eliminado a " + eliminado + ". ¡Menos drama!")
        else:
            print("IA: No hay nadie a quien eliminar.")

    elif opcion == "5":
        print("IA: Cerrando sesión. ¡Sigue brillando en las redes!")

    else:
        print("IA: Opción no válida, intenta de nuevo.")
        print("elaborado por karlasaenz0133")
```
